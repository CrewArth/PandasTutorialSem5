```python
import pandas as pd

df = pd.read_csv('Dataset.csv') #This will access the file and read the data.

df.head(10)
```

```python
#To Find the total Duplicate Records.

Duplicate_Records = df.duplicated(subset='customerID').sum()

print(Duplicate_Records)
```

```python
#To Find missing values of a variable.

Missing_Values = df['tenure'].isnull().sum()

print(Missing_Values)
```
